### What is a Boxplot?

For some datasets, additional information that measures of cntral tendancy like median, mean and mode is required. Analysts  need to know how data is dispersed. A boxplot is a method of displaying the distribution of data based on a five summay numbers
    - Minimum (excluding outliers)
    - First Quartile Q1 (the 25th Percentile) middle value between the smallest number (not min) and the median of dataset.
    - Median Q2 (50th Percentile) This is the middle of the dataset.
    - Third quartile Q3 (the 75th Percentile) middle value between the median and the highest value (not max).
    - Maximum (excluding outliers)
The interquartile range (IQR) lies between the 25th and the 75th percentile. A Boxplot is constructed by drawing a box between the upper and lower quartiles, drawing a solid line across the box to locate the median. Boxplots can highlight outliers (shown as green circles in above image) and are very good at showing how data is spread out. They have an advantage over other plot types such as histograms of taking up less space. This is useful when you want to compare data from two columns of a DataFrame.

***

### Quartile
Quartiles are the values that divide a list of numbers into quarters. 

For example, we can find the quartiles from the following list - (1, 2, 5, 6, 7, 9, 12, 15, 18, 19, 27).

Find the median = 9

Place parentheses around the numbers above and below the median. 
(1, 2, 5, 6, 7), 9, (12, 15, 18, 19, 27).

Find Q1 - the median of the lower half = 5

Find Q3 - the median of the upper half = 18


Subtract Q1 from Q3 to find the interquartile range IQR.
18 – 5 = 13.

### Outlier

An outlier is an observation that lies an abnormal distance from other 'normal' values. The analyst decides which values are classified as normal before abnormal observations can be identified.(1)The following quantities (called fences) are needed for identifying extreme values in the tails of the distribution: 
    - lower inner fence: Q1 - 1.5*IQ 
    - upper inner fence: Q3 + 1.5*IQ 
    - lower outer fence: Q1 - 3*IQ 
     -upper outer fence: Q3 + 3*IQ 
Outlier detection criteria 
A point beyond an inner fence on either side is considered a mild outlier. A point beyond an outer fence is considered an extreme outlier. I have referred to this (1) onlie information because it is well defined and easily explaind in a project setting.

### Boxplot - Sepal Length
    
For Iris data set Sepal Length values, the mean (5.843) is almost the same as the median (5.8) as seen in df.describe(). If we look at the below boxplot diagram, we can see that the diagram is almost symmetrical, meaning the mean value is almost the same as the median. The distribution is normal, meaning there are few exceptionally large or small values. The whiskers extend from the lowest value of 4.3 and highest value of 7.9. Lower quartile extends from left whisker at 4.3 to the begining of the interquartile range at 5.1. The interquartile range exrends from 5.1 to 6.4 and the upper quartile from 6.4 to 7.9. 
***

```python
from matplotlib import rcParams
sns.boxplot(data=df.sepal_length, orient="h").set_title('Iris Data - Sepal Length Boxplot') # horizontal boxplot
sns.set(rc={'figure.figsize':(8,2)})
plt.grid(True)
```

#### Historgam Vs Boxplot
Sepal lenght values are also used in the below displot graph. Quartile, median, lowest and highest values are easier to identify on boxplot diagram. However, histograms are better at describing the distribution of data. Histograms plot data with ranges of the data grouped into bins. One disadvantage of histograms is that they are not as readable as boxplot when comparing distributions across multiple categories. 

```python
sns.distplot(df.sepal_length, color="r").set_title('Iris Data - Sepal Length Distplot')
```

#### Density & Rugplot
A density plot is a smooth version of the histogram. Density plots calculate a distribution at each data point. It can generate data that falls outside the bounds of the origional data. This is not an issue with boxplots as the start and fiih points are well defined. Rugplots show all values of a distribution on x-axis, but is limited in other aspects of data visualisation. 

```python
sns.kdeplot(df.sepal_length, shade=True, color="y").set_title('Iris Data - Sepal Length Kdeplot');
sns.rugplot(df.sepal_length);
```

### Boxplot - Sepal Width 

#### Check for Outliers - lower fence 

For sepal width, median 3.0 and mean 3.05 values are similar, meaning plot is almost symmetrical. Outliers can be observed in the sepal width diagram. Outliers are identified outside the lower quartile by calculating quartile Q1 minus 1.5 times interquartile range IQR.

Lower fence => (Q1) 2.8 - (1.5 X (IQR) 0.5) = 2.05

There is one value less than lower fence value of 2.05. Index = 60, SepalWidth = 2.0. This is the lower outlier and can be seen on the left hand side of the lower whisker of the Sepal Width boxplot.

```python
check_lower_fence = df.loc[:,'sepal_width'] # filter sepal width data
check_lower_fence.loc[df.sepal_width < 2.05 ] # check f there are values less than lower fence value of 2.05
```

### Advantages of Boxplots
Boxplots have the advantage of taking up less space which can be useful when comparing distributions of different groupings of data. The Outliers, Percentiles, Median, Symmetry and spread of the data are easily identified. In the below 'Grouping Boxplot & Swarmplot' diagram, four different distributions are compared. Swarmplot enhances the visualisation of the data. For instance, for sepal width, we can see how tightly the data is grouped together. The opposite is true for petal length, where there is a greater spread of data. 

```python
from matplotlib import rcParams 

ax = sns.boxplot(data=df, orient="h").set_title('Grouping Boxplot & Swarmplot') # boxplot of four columns of data
ax = sns.swarmplot(data=df, orient="h") # swarmplot of four columns of data
rcParams['figure.figsize'] = 12,5

for patch in ax.artists:
 r, g, b, a = patch.get_facecolor()
 patch.set_facecolor((r, g, b, .5)) # Add transparency to colors so that swarmplot data can be seen
plt.grid(True)
```

### Pairgrid
